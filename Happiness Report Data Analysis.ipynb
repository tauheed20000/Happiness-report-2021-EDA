# Data Content


The happiness scores and rankings use data from the Gallup World Poll

Gallup World Poll: In 2005, Gallup began its World Poll, whic continually surveys citizens i 150 countries, reprezenting more than 98% of the world's adult population. The Gallup WOrld Poll consists of more than 100 global questions as well as region-specific items

.
The columns following the happiness score estimate the extent to which each of six factors-economic production, social support, life expactancy, freedom, absence the world's lowest national averages for each of the six factors.

# Ladder score:
Happiness score or subjective well-being. This is the national average response to the question of life evaluations..report/

# Logged GDP per capita:
The GDP-per-capita time series from 2019 to 2020 using countryspecific forecast of real GDP growth in 2020.
Social support: Social support refers to assistance or support provided by members of social networks to an individual


# Healthy life expentancy: 
Healthy life expectancy is the avarage life in good health-that is to say without irreversible limitation of activity in daily life or incapacities - of a fictitious generation subject to the conditions of morality an morbidity prevailing that year.


# Freedom to make life choices: 
Freedom to make life choices is the national avarage of binary responses to the GWP question "Are you satisfied or dissatisfied with your freedom to choose what do you with your life?" ... It is defined as the avarage of laughter and enjoyment for other waves where the happiness question was not asked

# Generosity:
Generosity is the residual of regressing national avarage response to the GWP questions "Have you donated money to a charity in the past month?" on GDP per capita

# Perceptions of corruption:
The measure is the national avarage of the survey responses to two questions in the GWP: "Is corruptions widespread throughout the goverment or not" and "Is corruption widespread within businesses or not?"
Ladder score in Dystopia: It has values equal to the world's lowest national averages. Dystopia as a bnechmark aganist whic to compare contributions from each of the six factors. Dystopia is an imaginary country that has the world's least-happy people. ... since life would be very unpleasant in a country with the world's lowest incomes, lowest life expantancy, lowest generosity, most corruption, least freedom and least social support, it is the referred to as "Dystopia," in contrast to Utopia

# plot between happiness and GDP

```python
plt.figure(figsize=(10, 6))
plt.title('Plot between happiness score and GDP')
sns.scatterplot(x='Ladder score', y='Logged GDP per capita', hue='Regional indicator', data=df)
plt.legend(loc='upper left', fontsize='6')
plt.xlabel('Ladder score or Happiness score')
plt.ylabel('GDP per capita')
plt.show()
```

```python
gdp_region = df.groupby('Regional indicator')['Logged GDP per capita'].sum()
```

```python
gdp_region
```

df.groupby('Regional indicator'): This part of the code is using the groupby method of a DataFrame. This method is used to group the DataFrame by a specified column, in this case, 'Regional indicator'. This means that the DataFrame will be divided into groups based on the unique values in the 'Regional indicator' column.

['Logged GDP per capita']: After using groupby, we specify the column we're interested in, which is 'Logged GDP per capita'. This is the column that we want to perform operations on within each group.

```python
gdp_region.plot.pie(autopct='%1.1f%%')
plt.title('GDP by Region')
plt.ylabel("")
plt.show()
```

autopct='%1.1f%%': This parameter formats the labels on the pie chart to display percentages with one decimal place.

```python
top_10 = df.head(10)
bottom_10 = df.tail(10)
```

```python
# Creating a bar plot for the top 10
plt.figure(figsize=(10, 6))
plt.bar(top_10['Country name'], top_10['Ladder score'], color='blue', label='Top 10')
plt.title('Top 10 happiest Countries by Ladder Score')
plt.xlabel('Country Name')
plt.ylabel('Ladder Score')
plt.xticks(rotation=45)
plt.legend()

# Creating a bar plot for the bottom 10
plt.figure(figsize=(10, 6))
plt.bar(bottom_10['Country name'], bottom_10['Ladder score'], color='red', label='Bottom 10')
plt.title('Bottom 10 happies Countries by Ladder Score')
plt.xlabel('Country Name')
plt.ylabel('Ladder Score')
plt.xticks(rotation=45)
plt.legend()

plt.show()
```
